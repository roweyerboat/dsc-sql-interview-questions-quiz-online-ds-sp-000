## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

_______________________________________________________________________________________________________________________________

Primary Keys are unique identifiers for the table.  There can only be unique entries in these columns.  Foreign keys are a primary key from a different table.  The relationship between them is you can match a primary key from one table to the foreign key of another in order to join the tables to line up the corresponding data.



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

There are 3 types of relationships entities that can exist in a SQL database: one-to-one, one-to-many, many-to-many.
One-to-one would mean a row in a table has exactly one matching row to in another table.
One-to-many would mean one row in one table is connected to many rows in another table.  For example, an employee could be a sales rep to many customers.
Many-to-many would mean many rows in one table pair with many rows in another table.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

INNER JOIN - All the records from the first table and the second table are joined where the join condition is met
LEFT JOIN - All the records from the first table along with records from the second table for which the join condition is met
RIGHT JOIN - All the records from the second table along with records from the first table for which the join condition is met
OUTER JOIN - All the records are join regardless if they have a match with the other table
